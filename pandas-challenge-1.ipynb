## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
df['line_subtotal'] = df['unit_price'] * df['qty']
df.head(2)
```

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
df['line_subtotal'] = df['unit_price'] * df['qty']
line_subtotal_df = df[[ 'unit_price', 'qty','line_subtotal']]
line_subtotal_df.head(2)
```

```python
df['total_weight'] = df['unit_weight'] * df['qty']
df.head(2)
```

```python
def cal_shipping_price(total_weight):
    if total_weight > 50:
        return total_weight * 7  
    else:
        return total_weight * 10  
```

```python
df['shipping_price'] = df['total_weight'].apply(cal_shipping_price)
df.head(2)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.

shipping_price_df = df[['unit_price', 'unit_weight' , 'qty' , 'total_weight', 'shipping_price']]
shipping_price_df.head(3)
```

```python
df['total_price'] = df['line_subtotal'] + df['shipping_price']
total_price_df = df[['unit_price', 'unit_weight' , 'qty' , 'total_weight', 'shipping_price', 'total_price']]
total_price_df.head(3)
```

```python
sales_taxes = 0.0925

after_taxes = df['total_price'] * sales_taxes
df['line_price'] = df['total_price'] + after_taxes
df['line_price'] = df['line_price'].round(2)
```

```python
line_price_df = df[['unit_price', 'unit_weight' , 'qty' , 'total_weight', 'shipping_price', 'total_price', 'line_price']]
line_price_df.head(3)
```

```python
df['line_price'] = df['line_price'].round(2)

print(df[['line_price']])
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%  
line_price_df = df[['unit_price', 'unit_weight' , 'qty' , 'total_weight', 'shipping_price', 'total_price', 'line_price']]
line_price_df.head(3)
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
df['line_cost'] = df['unit_cost'] * df['qty'] + df['shipping_price']
df.head(3)
```

```python
# Create a column for the profit of
# each line using line cost and line price
df['line_profit'] = df['line_price']- df['line_cost']
df.head(3)
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
top5_clients_df = ["33615","66037","46820","38378","24741"]
```

```python
top5_clients_df = [int(client_id) for client_id in top5_clients_df]
top5_clients_df
```

```python
filtered_df = df[df['client_id'].isin(top5_clients_df)]
filtered_df
```

```python
line_profit_per_client = filtered_df.groupby('client_id')['line_profit'].sum()
line_profit_per_client
for client_id, line_profit in line_profit_per_client.items():
    print(f"{client_id}: ${line_profit:.2f}")
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.

summary_filtered_df_sorted = summary_filtered_df.sort_values(by='line_profit', ascending=False)
summary_filtered_df_sorted
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.

summary_filtered_df_sorted.rename(index={"qty": "Quantity","shipping_price":"Shipping","line_price":"Total Revenue",
                                        "line_cost":"Total Cost","line_profit":"Total Profit"})
summary_filtered_df_sorted
```

```python
# Sort the updated data by "Total Profit" form highest to lowest
TotalProfit_filtered_df_sorted = summary_filtered_df.sort_values(by='line_profit', ascending=False)
TotalProfit_filtered_df_sorted
```
